```python
# Game-Z

def Zro1(): # U-U
    weight = a14*Z+b14 + (a5*(x0+z0)+b5)+b13 + (a12*(x3+y3+z3)+b12)
    return(weight)

def Zro2(): # U-D
    weight = a14*Z+b14 + (a5*(x0+z0)+b5)+b13 + ((a15+a16)*(y2+z2)+b15+b16)
    return(weight)

def Zro3(): # D-U
    weight = a14*Z+b14 + (a11*(x1+y0+z1)+b11) + (a12*(x3+y3+z3)+b12)
    return(weight)

def Zro4(): # D-D
    weight = a14*Z+b14 + (a11*(x1+y0+z1)+b11) + ((a15+a16)*(y2+z2)+b15+b16)
    return(weight)
```
